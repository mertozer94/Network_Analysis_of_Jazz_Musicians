### 4.3 Histogram of distances

```python
#path = nx.all_pairs_shortest_path_length(G)
#random_path = nx.all_pairs_shortest_path_length(random_G)
```

```python
def getShortestDegree(G):
    shortest_degree = {}
    visited = []
    path = nx.all_pairs_shortest_path_length(G)
    for i in path:
        for key, value in path.get(i).items():
            if value in shortest_degree:
                key_value = str(key) + str(value)
                if not(key_value in visited):
                    shortest_degree[value]+= 1
            else:
                shortest_degree[value] = 1
            
            value_key = str(value) + str(key)
            visited.append(value_key)
        
    return shortest_degree
```

```python
def toArray(dictionnary):
    dictlist = []
    for key, value in dictionnary.items():
            
        dictlist.append(value)
    return dictlist
```

```python
shortest_degree = getShortestDegree(G)
random_shortest_degree = getShortestDegree(random_G)

shortest_degree_list = []
random_shortest_degree_list = []

shortest_degree_list = toArray(shortest_degree)
random_shortest_degree_list = toArray(random_shortest_degree)
```

```python
deg_seq = [x for x in shortest_degree_list]
plt.hist(deg_seq)
plt.title('Degree Histogram of shortest degree on jazz musicians dataset')
plt.xlabel('degree')
plt.ylabel('frequency')
plt.show()
```

```python
deg_seq = [x for x in random_shortest_degree_list]
plt.hist(deg_seq)
plt.title('Degree Histogram of shortest degree on random dataset')
plt.xlabel('degree')
plt.ylabel('frequency')
plt.show()
```

## 5. Detect the communities in the graph, and discuss the results.

```python
import community

def computeCommunity(G):
    #first compute the best partition
    partition = community.best_partition(G)

    #drawing
    size = float(len(set(partition.values())))
    pos = nx.spring_layout(G)
    count = 0.
    for com in set(partition.values()) :
        count = count + 1.
        list_nodes = [nodes for nodes in partition.keys()
                                    if partition[nodes] == com]
        nx.draw_networkx_nodes(G, pos, list_nodes, node_size = 20,
                                    node_color = str(count / size))


    nx.draw_networkx_edges(G,pos, alpha=0.5)
    plt.show()
```

```python
computeCommunity(G)
```

```python
computeCommunity(random_G)
```
